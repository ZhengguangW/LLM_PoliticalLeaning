```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
# Initialize Selenium WebDriver
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url = "https://www.thedailybeast.com/category/politics"
print("yes")
driver.get(url)
print("yes")

# Function to count occurrences of a specific element
def count_occurrences():
    return len(driver.find_elements(By.CSS_SELECTOR, ".GridStory__title-link._15zfx"))

# Set a target count for the specific element
target_count = 250

# Counter for the specific element occurrences
occurrences_count = 0

# Loop to click the "Load More" button and count occurrences
while occurrences_count < target_count:
        # Click the "Load More" button
    load_more_button = driver.find_element(By.XPATH, "/html/body/div[2]/main/section/div/div[7]/button")
    driver.execute_script("arguments[0].click();", load_more_button)
    print("got")

    # Update the occurrences count
    occurrences_count = count_occurrences()
    
    print(f"Occurrences count: {occurrences_count}")

a_elements=driver.find_elements(By.CSS_SELECTOR, ".GridStory__title-link._15zfx a")
link = []
headline=[]
for a in a_elements:
    # Extract href of <a>
    href = a.get_attribute("href")
    link.append(href)
    # Find the <h3> element within each <a> and extract its text
    h3_text = a.find_element(By.CSS_SELECTOR, "h3").text
    headline.append(h3_text)
print(headline)
print(link)

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd 
DailyBeast=pd.DataFrame({
    "headline":headline,
    "headline-link":link
})
DailyBeast
```

```python
DailyBeast.to_csv("DailyBeast.csv")
```
