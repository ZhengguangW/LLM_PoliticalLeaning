```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd 
import time
```

```python
headline=[]
headline_link=[]
topic=[]
date=[]
```

```python


driver = webdriver.Chrome() 
driver.get("https://www.allsides.com/headline-roundups")
while True:
    headline_elements = WebDriverWait(driver, 120).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'td.views-field.views-field-name > a'))
    )
    topic_elements = WebDriverWait(driver, 120).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'td.views-field.views-field-field-story-topic'))
    )
    date_elements=WebDriverWait(driver, 120).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "td.views-field.views-field-field-story-date > span"))
    )
    for headline_element in headline_elements:
        reference = headline_element.get_attribute('href')
        print(reference)
        headline_link.append(reference)
        headline.append(headline_element.text)
    for topic_element in topic_elements:
        try:
            child=topic_element.find_element(By.CSS_SELECTOR,"a")
            topic.append(child.text)
            print(child.text)
        except:
            print("N/A")
            topic.append("N/A")
    for date_element in date_elements:
        date.append(date_element.text)
    next_button = WebDriverWait(driver, 120).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, 'ul > li.pager-next > a')))
    print(driver.current_url)
    next_button_url = next_button.get_attribute('href')
    driver.get(next_button_url)
    time.sleep(10)
    
```

```python
for _ in range (7):
    date.append(None)
```

```python
Allsides=pd.DataFrame()
Allsides["headline"]=headline
Allsides["headline-link"]=headline_link
Allsides["topic"]=topic
Allsides["date"]=date
```

```python
Allsides
Allsides.to_csv("Allsides-roundup.csv")
```

```python
import requests
from bs4 import BeautifulSoup
def GetIntoEachSummary(row):
    url=row["headline-link"]
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    left_element = soup.select_one('#block-views-story-id-boxs-block > div > div > div > div.news-item.left > a.news-title')
    central_element=soup.select_one('#block-views-story-id-boxs-block > div > div > div > div.news-item.center > a.news-title')
    right_element=soup.select_one('#block-views-story-id-boxs-block > div > div > div > div.news-item.right > a.news-title')
    if left_element:
        left_headline = left_element.get_text()
        left_link_allsides = left_element.get('href')
        left_link_original = soup.select_one('#block-views-story-id-boxs-block > div > div > div > div.news-item.left > div.flex.wrap.space-between > a').get('href')
    else:
        left_headline, left_link_allsides, left_link_original = None, None, None
    if central_element:
        central_headline = central_element.get_text()
        central_link_allsides = central_element.get('href')
        central_link_original = soup.select_one('#block-views-story-id-boxs-block > div > div > div > div.news-item.center > div.flex.wrap.space-between > a').get('href')
    else:
        central_headline, central_link_allsides, central_link_original = None, None, None
    if right_element:
        right_headline = right_element.get_text()
        right_link_allsides = right_element.get('href')
        right_link_original = soup.select_one('#block-views-story-id-boxs-block > div > div > div > div.news-item.right > div.flex.wrap.space-between > a').get('href')
    else:
        right_headline, right_link_allsides, right_link_original = None, None, None
    row['Left_Headline'] = left_headline
    row['Left_Link_Allsides'] = left_link_allsides
    row['Left_Link_Original'] = left_link_original
    row['Central_Headline'] = central_headline
    row['Central_Link_Allsides'] = central_link_allsides
    row['Central_Link_Original'] = central_link_original
    row['Right_Headline'] = right_headline
    row['Right_Link_Allsides'] = right_link_allsides
    row['Right_Link_Original'] = right_link_original
    return row

result_df = Allsides.apply(GetIntoEachSummary, axis=1)
result_df
```
