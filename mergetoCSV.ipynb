```python
import json
all_template=[]
all_template.append("headline")
for i in range (1,15):
    file_name=f'prompt{i}/Left.json'
    with open(file_name, 'r') as file:
        template=json.load(file)["template"]
    all_template.append(template)
    all_keys=[]
    for stance in ["Left","Right"]:
        file_name=f'prompt{i}/{stance}.json'
        with open(file_name, 'r') as file:
            data = json.load(file)["headline and yes-percentage"]
            for key,value in data.items():
                all_keys.append(key)
```

```python
print(all_template)
print(len(all_keys))
```

```python
import re
merged_data=[]
for key in all_keys:
    pattern = r'Statement: (.*?). Question'
    key = re.findall(pattern, key)[0]
    row=[]
    row.append(key)
    for i in range (1,15):
        for stance in ["Left","Right"]:
            file_name=f'prompt{i}/{stance}.json'
            with open(file_name, 'r') as file:
                data=json.load(file)
                template=data["template"]
                data1=data["headline and yes-percentage"]
                try:
                    key_final=template.replace("<Statement>",key)
                    row.append(data1[key_final])
                except:
                    pass
    merged_data.append(row)
    
```

```python
merged_data
```

```python
import pandas as pd 
df = pd.DataFrame(merged_data, columns=all_template)

df.to_excel("merged4.xlsx")
```

```python
import pandas as pd
from scipy.stats import spearmanr
df=pd.read_excel("merged4.xlsx")
nan_check = df.isna()

# Check if there are any NaN values in the DataFrame
any_nan = nan_check.any().any()

print("Does the DataFrame contain NaN values?")
print(any_nan)
rho, p_value = spearmanr(df.iloc[:,2],df.iloc[:,3])
print(rho,p_value)
```

```python
filtered=df.iloc[:,2:16]
filtered
```

```python
correlation_matrix=[]
for i in range (0,14):
    correlation_row=[]
    for j in range (0,14):
        rho, p_value = spearmanr(filtered.iloc[:,i],filtered.iloc[:,j])
        correlation_row.append(rho)
    correlation_matrix.append(correlation_row)
correlation_matrix

rcnames=filtered.columns.to_list()
correlation_df = pd.DataFrame(correlation_matrix,index=rcnames,columns=rcnames)
correlation_df.to_csv("spearman4.csv")
        
```
